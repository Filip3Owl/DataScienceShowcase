Some uses of SciPy.stats lib:

```python
# The median of a standard Student's t distribution with 10 degrees of freedom
print(stats.t(10).ppf(0.5))

# The probability of getting exactly 3 heads in 10 flips of a fair coin
print(stats.binom(10, 0.5).pmf(3))
```

MatPlot lib cases:

```python
x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)
# Plotting the sine function
plt.plot(x, y)
plt.title('Sine Function')
```

Fancier version:

```python
# Display a grid on the plot for better readability.
plt.grid(True)
# Plot the 'x' and 'y' data with a dotted blue line, 5 units wide, and semi-transparent.
plt.plot(x, y, ":", lw=5, color='blue', alpha=0.5)
# Set the label for the y-axis.
plt.ylabel('sin(x)', size=20)
# Set the label for the x-axis.
plt.xlabel('x', size=20)
```

```python
# Create an array 'x' ranging from 0 to 3*pi with a step of 0.01.
x = np.arange(0, 3 * np.pi, 0.01)
# Calculate the sine of each value in 'x'.
y_sin = np.sin(x)
# Calculate the cosine of each value in 'x'.
y_cos = np.cos(x)

# Plot y_sin against x.
plt.plot(x, y_sin)
# Plot y_cos against x.
plt.plot(x, y_cos)
# Set the label for the y-axis.
plt.ylabel('x axis label', size=20)
# Set the label for the x-axis.
plt.xlabel('y axis label', size=20)
# Set the title of the plot.
plt.title('Sine and Cosine Functions')
# Add a legend to differentiate the lines.
plt.legend(['sin(x)', 'cos(x)'])
# Display a grid on the plot.
plt.grid(True)
```

```python
# Create an array 'x' ranging from 0 to 3*pi with a step of 0.01.
x = np.arange(0, 3 * np.pi, 0.01)
# Calculate the sine of each value in 'x'.
y_sin = np.sin(x)
# Calculate the cosine of each value in 'x'.
x_cos = np.cos(x) # Note: This was likely meant to be y_cos for consistency.

# Create a subplot grid (2 rows, 1 column) and select the first subplot.
plt.subplot(2, 1, 1)
# Plot the sine function.
plt.plot(x, y_sin)
# Set the title for the sine plot.
plt.title('Sine Function')

# Select the second subplot in the grid.
plt.subplot(2, 1, 2)
# Plot the cosine function.
plt.plot(x, x_cos) # Note: Assuming x_cos was intended to be plotted here.
# Set the title for the cosine plot.
plt.title('Cosine Function')
# Adjust subplot parameters for a tight layout, preventing labels/titles from overlapping.
plt.tight_layout()
```

Seaborn:

```python
# Define the file path for the dataset.
path = r"C:\Users\Felipe\Documents\Filipe docs\ai projects\DataScienceShowcase\Datasets\cartwheel.csv"
# Load the CSV file into a pandas DataFrame.
df = pd.read_csv(path)
# Create a scatter plot using seaborn, showing 'Wingspan' on the x-axis and 'CWDistance' on the y-axis.
sns.scatterplot(data=df, x='Wingspan', y='CWDistance')
```

```python
# Points are colored differently based on the 'Gender' column, allowing for group comparison.
sns.scatterplot(x="Wingspan", y="CWDistance", hue="Gender", data=df)
```

```python
# This plot shows individual data points and helps visualize density and spread without overlap.
sns.swarmplot(x="Gender", y="CWDistance", data=df)
```

```python
# This visualizes the distribution, median, quartiles, and outliers for each chosen variable.
sns.boxplot(data=df.loc[:, ["Age", "Height", "Wingspan", "CWDistance", "Score"]])
```

```python
# A boxplot only of females
sns.boxplot(data=df.loc[df['Gender'] == 'F', ["Age", "Height", "Wingspan", "CWDistance", "Score"]])
```

```python
# A boxplot only of males
sns.boxplot(data=df.loc[df['Gender'] == 'M', ["Age", "Height", "Wingspan", "CWDistance", "Score"]])
```

Histogram:

```python
# 'kde=True' adds a Kernel Density Estimate line to show the probability density function.
# 'rug=True' adds a rug plot, which draws small vertical lines at each data point on the x-axis.
# 'color='Green'' sets the color of the plot elements to green.
# 'height=5' sets the height of the plot in inches.
# 'aspect=1.5' sets the aspect ratio of the plot (width = height * aspect).
sns.displot(df.CWDistance, kde=True, rug=True, color='Green', height=5, aspect=1.5)
```

Countplot:

```python
# Create a count plot using seaborn, showing the number of occurrences for each 'Gender' category.
sns.countplot(x="Gender", data=df)
# Rotate the x-axis labels to 0 degrees (horizontal) for better readability if labels overlap.
plt.xticks(rotation=45)
```
