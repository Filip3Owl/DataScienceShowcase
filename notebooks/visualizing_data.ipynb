Tables, histograms and boxplots

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

Importing dataset:

```python
tips_data = sns.load_dataset("tips")
```

```python
tips_data.head()
```

```python
tips_data.describe()
```

Capturing the shape of the distribution:

```python
sns.histplot(tips_data['total_bill'], kde=False).set_title('Histogram of Total Bill Amounts')
```

```python
sns.histplot(tips_data['tip'], kde=False).set_title('Histogram of Tips Amounts')
```

We have bivariate data. Let's plot a scatter plot:

```python
sns.scatterplot(x='total_bill', y='tip', data=tips_data).set_title('Scatter Plot of Total Bill vs Tip')
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=tips_data).set_title('Scatter Plot of Total Bill vs Tip')
```

Histograms need tons of data to show a clear picture of its distribution. **Boxplots**, however, are great with **less data**, offering a quick summary of a distribution's key features.

```python
sns.boxplot(x=tips_data['total_bill']).set_title('Box Plot of Total Bill Amounts')
```

```python
sns.boxplot(x=tips_data['tip']).set_title('Box Plot of Tips Amounts')
```

Stratified by groups:

```python
sns.boxplot(x=tips_data["tip"], y=tips_data["smoker"]).set_title('Box Plot of Tips by Smoker Status')
```

```python
sns.boxplot(x=tips_data["tip"], y=tips_data["time"]).set_title('Box Plot of Tips by Time of Day')
```

```python
b = sns.FacetGrid(tips_data, row="time")
b = b.map(plt.hist, "tip")
plt.ylabel('Frequency')
plt.xlabel('Tip Amount')
```

```python
sns.boxplot(x=tips_data['tip'], y=tips_data['day']).set_title('Box Plot of Tips by Day of the Week')
```

```python
h = sns.FacetGrid(tips_data, row="day")
h = h.map(plt.hist, "tip")
plt.xlabel('Tip Amount')
plt.ylabel('Frequency')
plt.show()
```
